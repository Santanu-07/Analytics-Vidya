```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
train = pd.read_csv(r"C:\Users\Santanu Das\Desktop\New folder (2)\train_HK6lq50.csv")
test = pd.read_csv(r"C:\Users\Santanu Das\Desktop\New folder (2)\test_wF0Ps6O.csv")
```

```python
train.head()
```

```python
train.describe()
```

```python
train['age'] = train['age'].fillna(train['age'].median())
```

```python
train.head()
```

```python
train = train.drop(['id'], axis=1)
train.head()
```

```python
pd.unique(train.program_id)
```

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder() 
```

```python
train['program_id']= le.fit_transform(train['program_id'])
train['program_type']= le.fit_transform(train['program_type'])
train['trainee_id']= le.fit_transform(train['trainee_id'])
train['difficulty_level']= le.fit_transform(train['difficulty_level'])
train['education']= le.fit_transform(train['education'])
train['program_id'].unique()
train['program_type'].unique()
train['trainee_id'].unique()
train['difficulty_level'].unique()
train['education'].unique()
```

```python
train.head()
```

```python
train = train.assign(gender=(train['gender'] == 'M').astype(int))
train = train.assign(test_type=(train['test_type'] == 'online').astype(int))
train = train.assign(is_handicapped=(train['is_handicapped'] == 'Y').astype(int))
train.head()
```

```python
train.shape
```

```python
X = train.drop(['is_pass'], axis=1)
y = train['is_pass']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.01)
```

```python
from lightgbm import LGBMClassifier
```

```python
model = LGBMClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn import metrics 
metrics.accuracy_score(y_test, y_pred)*100
```
